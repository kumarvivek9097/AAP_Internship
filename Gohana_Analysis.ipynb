```python
data1=pd.read_csv('/home/vivek/Documents/Internship/AAP_Data_Analytics/Task 6/Part 2/gohana/2005_VOTE%.csv')
data1.index = np.arange(1, len(data1)+1)
data1.drop(['Sr. No.'], axis = 1,inplace=True)
data1.tail()
```

```python
dt=data1.idxmax(axis=1)
count=0
count1=0
count2=0
count3=0
count4=0
count5=0
for i in range(len(dt)):
    if dt[i+1]=='inc %':
        count=count+1
    elif dt[i+1]=='inld %':
        count1=count1+1
    elif dt[i+1]=='bjp %':
        count2=count2+1
    elif dt[i+1]=='jd(s) %':
        count3=count3+1
    elif dt[i+1]=='bsp %':
        count4=count4+1
    elif dt[i+1]=='others %':
        count5=count5+1
print(count)
print(count1)
print(count2)
print(count3)
print(count4)
print(count5)
```
