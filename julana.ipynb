```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('/home/vivek/Documents/Internship/AAP_Data_Analytics/Task 5/JULANA - Sheet2.csv')
data.drop(['VOTES'],axis=1,inplace=True)
data.head()
```

```python
df=data.loc[(data['YEAR']==2005)]
df1=df.sort_values(by=['VOTES PERCENTAGE'], ascending=False)
df2=df1[['VOTES PERCENTAGE']]
df2 = df2.reset_index(drop=True)
#for i in range(0,15):
#    df3=df2.iloc[i]-df2.iloc[i+1]
#    print df3
```

```python
df=data.loc[(data['YEAR']==2014)]
df1=df.sort_values(by=['VOTES PERCENTAGE'], ascending=False).head(6)
x=df1['PARTY']
y=df1['VOTES PERCENTAGE']
plt.figure(figsize=(20,10))
plt.grid(True, zorder=5)
plt.xlabel('Party',fontsize=20)
plt.ylabel('Votes Percentage',fontsize=20)
plt.title('2014 (Julana)',fontsize=20)
plt.tick_params(labelsize=15)
plt.bar(x,y)
```

```python
print(x)
print(y)
print(100-y.sum())
```

```python
#figure = plt.figure()
labels=['INLD','INC','BSP','BJP','CPI','HJC(BL)','OTHERS']
size=[43.962152,
25.642908,
17.957860,
9.683951,
0.793753,0.466249,1.4931272617000104]
plt.figure(figsize=(20,10))
figure,ax1 = plt.subplots()
colors = ['blue','green','grey','orange','black','red','pink']
result=ax1.pie(size,labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90,colors=colors)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
fig = plt.gcf()
fig.set_size_inches(10,10)
plt.legend(result[0], size,  bbox_to_anchor=(0.00, 1))
plt.title('2014 (Julana)',fontsize=20)
figure.subplots_adjust(top=0.95)
plt.show()
```
