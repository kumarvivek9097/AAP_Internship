```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('/home/vivek/Downloads/GOHANA - Sheet1(1).csv')
data.drop(['TOTAL VOTES POLLED'],axis=1,inplace=True)
data.head()
```

```python
df=data.loc[(data['PARTY']=='BJP')]
X1=df['YEAR']
Y1=df['VOTES PERCENTAGE']
df1=data.loc[(data['PARTY']=='INC')]
X2=df1['YEAR']
Y2=df1['VOTES PERCENTAGE']
df2=data.loc[(data['PARTY']=='INLD')]
X3=df2['YEAR']
Y3=df2['VOTES PERCENTAGE']
df3=data.loc[(data['PARTY']=='BSP')]
X4=df3['YEAR']
Y4=df3['VOTES PERCENTAGE']
df4=data.loc[(data['PARTY']=='SAP')]
X5=df4['YEAR']
Y5=df4['VOTES PERCENTAGE']
df5=data.loc[(data['PARTY']=='HVP')]
X6=df5['YEAR']
Y6=df5['VOTES PERCENTAGE']
X7=data['YEAR']
Y7=data['Literacy Rate']
plt.bar(X7,Y7)
```

```python
plt.figure(figsize=(20,10))
ax = plt.subplot(111)
ax.bar(X7-0.6, Y7,width=0.2,color='c',align='center')
ax.bar(X1-0.4, Y1,width=0.2,color='b',align='center')
ax.bar(X2-0.2, Y2,width=0.2,color='g',align='center')
ax.bar(X3, Y3,width=0.2,color='r',align='center')
ax.bar(X4+0.2, Y4,width=0.2,color='black',align='center')
ax.bar(X5+0.4, Y5,width=0.2,color='yellow',align='center')
ax.bar(X6+0.6, Y6,width=0.2,color='PINK',align='center')
#ax.xaxis_date()
ax.legend(["Literacy Rate","BJP", "INC","INLD","BSP","SAP","HVP"],fontsize=20)
plt.xlabel('Year',fontsize=20)
plt.ylabel('Votes Percentage',fontsize=20)
plt.show()
```
