```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('/home/vivek/Documents/Internship/AAP_Data_Analytics/Task 5/BARODA - Sheet1(2).csv')
data.drop(['TOTAL VOTES POLLED'],axis=1,inplace=True)
data.head()
```

```python
df=data.loc[(data['PARTY']=='BJP')]
X1=df['YEAR']
Y1=df['% VOTES']
plt.figure(figsize=(20,10))
plt.grid(True, zorder=5)
plt.xlabel('Year',fontsize=20)
plt.ylabel('Votes Percentage',fontsize=20)
plt.title('BJP (Baroda)',fontsize=20)
plt.bar(X1,Y1)
```
